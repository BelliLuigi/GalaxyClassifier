```python
import pickle
import io
import builtins

class GalaxyNet():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"GalaxyNet(name={self.name}, age={self.age})"

class SafeUnpickler(pickle.Unpickler):
    safe_builtins = {
        'builtins': {
            'dict': dict,
            'list': list,
            'str': str,
            'int': int,
            'float': float,
            'set': set,
            'tuple': tuple,
            'bool': bool,
            'NoneType': type(None),
        }
    }

    def find_class(self, module, name):
        if module in self.safe_builtins and name in self.safe_builtins[module]:
            return self.safe_builtins[module][name]
        raise pickle.UnpicklingError(f"Unauthorized class: {module}.{name}")

def safe_load(file_path):
    with open(file_path, 'rb') as f:
        return SafeUnpickler(f).load()

# Usage
data = safe_load('blob')
print(data)
```
